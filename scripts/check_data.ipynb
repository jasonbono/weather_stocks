```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
basedir = '/Users/bono/demo_projects/weather_stocks/'
s_file = basedir+'data/df_stocks.pkl'
w_file = basedir+'data/df_weather.pkl'
s = open(s_file,'rb')
w = open(w_file,'rb')
df_stocks = pickle.load(s)
df_weather = pickle.load(w)
s.close()
w.close()


c_file = basedir+'data/df_clean.pkl'
a_file = basedir+'data/df_all.pkl'
c = open(c_file,'rb')
a = open(a_file,'rb')
df_all = pickle.load(a)
df_clean = pickle.load(c)
c.close()
a.close
```

```python
df_clean['temp_date'] = df_clean.index
df_all['temp_date'] = df_all.index
df_weather['temp_date'] = df_weather.index
df_stocks['temp_date'] = df_stocks.index
```

```python
df_weather['SNOW'].fillna(0)
```

```python
#by day of month
plt.figure(figsize=(15, 6))
df_weather['SNOW'].groupby(df_weather['temp_date'].dt.day).count().plot(kind="bar")
plt.figure(figsize=(15, 6))
df_stocks['Open'].groupby(df_stocks['temp_date'].dt.day).count().plot(kind="bar")
plt.figure(figsize=(15, 6))
df_clean['SNOW'].groupby(df_clean['temp_date'].dt.day).count().plot(kind="bar")
plt.figure(figsize=(15, 6))
df_all['SNOW'].groupby(df_all['temp_date'].dt.day).count().plot(kind="bar")
```

```python
#by month
plt.figure(figsize=(15, 6))
df_weather['SNOW'].groupby(df_weather['temp_date'].dt.month).count().plot(kind="bar")
plt.figure(figsize=(15, 6))
df_stocks['Open'].groupby(df_stocks['temp_date'].dt.month).count().plot(kind="bar")
plt.figure(figsize=(15, 6))
df_clean['SNOW'].groupby(df_clean['temp_date'].dt.month).count().plot(kind="bar")
plt.figure(figsize=(15, 6))
df_all['SNOW'].groupby(df_all['temp_date'].dt.month).count().plot(kind="bar")
```

```python
#by year
plt.figure(figsize=(15, 6))
df_weather['SNOW'].groupby(df_weather['temp_date'].dt.year).count().plot(kind="bar")
plt.figure(figsize=(15, 6))
df_stocks['Open'].groupby(df_stocks['temp_date'].dt.year).count().plot(kind="bar")
plt.figure(figsize=(15, 6))
df_clean['SNOW'].groupby(df_clean['temp_date'].dt.year).count().plot(kind="bar")
plt.figure(figsize=(15, 6))
df_all['SNOW'].groupby(df_all['temp_date'].dt.year).count().plot(kind="bar")
```
