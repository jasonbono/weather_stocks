# Get the stock data

This script scrapes the data from yahoo finance and stores it as a dataframe in a pickel file for further analysis elsewhere. It is made very easy with yfanance

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = yf.download("DJIA", start="1950-01-01", end="2020-01-01",
                   group_by="ticker")
#rename the index to match the weather's index
df.index.names = ['DATE']
df.index = pd.to_datetime(df.index)
```

```python
df.shape
#safe the df as a pickle file
basedir = '/Users/bono/demo_projects/weather_stocks/'
file = basedir+'data/df_stocks.pkl'
df.to_pickle(file)
```

# Optional: Make sure the distribution of observations is roughly uniform

```python
df['temp_date'] = df.index
```

```python
#by day of month
plt.figure(figsize=(15, 6))
df['Open'].groupby(df['temp_date'].dt.day).count().plot(kind="bar")
```

```python
#by month
plt.figure(figsize=(15, 6))
df['Open'].groupby(df['temp_date'].dt.month).count().plot(kind="bar")
```

```python
#by year
plt.figure(figsize=(16, 6))
df['Open'].groupby(df['temp_date'].dt.year).count().plot(kind="bar")
```
