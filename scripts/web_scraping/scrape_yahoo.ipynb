# Get the stock data

This script scrapes the data from yahoo finance and stores it as a dataframe in a pickel file for further analysis elsewhere. It is made very easy with yfanance

### load what's needed

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

### get the stock data into a data frame

```python
df = yf.download("DJIA", start="1950-01-01", end="2020-01-01",
                   group_by="ticker")
#rename the index to match the weather's index
df.index.names = ['DATE']
df.index = pd.to_datetime(df.index)
```

### save the df as a pickle file

```python
basedir = '/Users/bono/demo_projects/weather_stocks/'
file = basedir+'data/df_stocks.pkl'
df.to_pickle(file)
```

# Optional: Make sure the distribution of observations is roughly uniform

### Plot the number of observations for each day of the month

```python
plt.figure(figsize=(15, 6))
df['Open'].groupby(df.index.day).count().plot(kind="bar")
```

### Plot the number of observations for each month of the year

```python
plt.figure(figsize=(15, 6))
df['Open'].groupby(df.index.month).count().plot(kind="bar")
```

### Plot the number of observations for each year

```python
plt.figure(figsize=(16, 6))
df['Open'].groupby(df.index.year).count().plot(kind="bar")
```

looks good. expect ~253 trading days per year, with near uniformity
